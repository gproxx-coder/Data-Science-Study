### Here Basic columns are 10 But because of Dummies the SOO many categorical variables are getting created So its not efficient.

### Instead of doing that way we will select 10 Categories which are having most numbers of occurances in the respectively Column (Variable). so only Maximum 10 columns can get selected.

```python
# df.X1.value_counts().head(10)
df.X2.value_counts().head(10)
```

```python
top_10_labels = [lab for lab in df.X2.value_counts().head(10).index]
top_10_labels
```

```python
def get_top_10_labels(df, variable):
#     top_10_labels = [label for label in df[variable].value_counts().head(10).index]
    top_10_labels = [lab for lab in variable.value_counts().head(10).index]
    return top_10_labels
```

```python
def one_hot_encoding(df, variable, top_10_labels):
    
    for lab in top_10_labels:
        df[variable+'_'+lab] = np.where(df[variable]==lab, 1, 0)
```

```python
top_10_labels = get_top_10_labels(df, df['X3'])
top_10_labels
```

```python
one_hot_encoding(df, 'X3', top_10_labels)
```
